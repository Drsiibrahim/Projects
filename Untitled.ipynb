```python
def outlier_lim(col):
    '''this function shows the upper limit and lower limit of a boxplot for each numerical column
    input: Numerical columns of a dataframe
    output: upper limit, lowe limit'''

    q3, q1 = np.percentile(col, [75, 25])
    iqr = q3 - q1
    ul = q3 + 1.5 * iqr
    ll = q3 - 1.5 * iqr

    return ul, ll
```

```python
def extract_feat(Address):
    return Address.split(',')[:3]

df1['city'] = df1['purchaseAddress'].apply(lambda x: extract_feat(x)[1])
df1['postalCode'] = df1['purchaseAddress'].apply(lambda x: extract_feat(x)[2])

```

```python
df1
```

```python
df1['orderDate'] = pd.to_datetime(df['orderDate'])
```

```python
df1.dtypes
```

```python
df1['manufacturerWarranty'] = df1['orderDate'] + pd.DateOffset(months = 6)
df1
```

```python
df1['orderTotal'] = df1['qtyOrdered'] * df1['priceEach']
df1
```

```python
df1['orderProfit'] = round(0.25 * df1['orderTotal'], 2)
df1
    
```

```python
df1_2019 = df1[df1['orderDate'].dt.year == 2019]
```

```python
df1_2019
```

```python
df1['orderDate'].unique()
```

```python
monthlyData = df1_2019.resample('M', on = 'orderDate').agg({
    'qtyOrdered':'sum',
    'priceEach':'sum',
    'orderTotal':'sum',
    'orderProfit':'sum'
})
```

```python
monthlyData
```

```python
monthlyData['totalSales'] = monthlyData['qtyOrdered'] * monthlyData['priceEach']
monthlyData
```

```python
monthlyData['totalSales'] = pd.to_numeric(monthlyData['totalSales'])
monthlyData
```

```python
fig, ax1 = plt.subplots(figsize = (10, 6))

ax1.plot(monthlyData.index.month_name(), monthlyData['totalSales'], color = 'blue', label = 'Total Sales')
ax1.set_xlabel('Month')
ax1.set_ylabel('Total Sales', color = 'blue')
ax1.tick_params(axis= 'y', labelcolor = 'blue')
ax1.yaxis.set_major_formatter('{:.2f}'.format)

ax2 = ax1.twinx()
ax2.plot(monthlyData.index.month_name(), monthlyData['orderProfit'], color = 'green', label = 'Total Profit')
ax2.set_ylabel('Total Profit', color = 'green')
ax2.tick_params(axis= 'y', labelcolor = 'green')
ax2.yaxis.set_major_formatter('{:.2f}'.format)


plt.title('Total Sales and Profit by Month (2019)')
#lines1, labels1 = ax1.get_legend_handles_labels()
#lines2, labels2 = ax2.get_legend_handles_labels()
plt.legend(lines1 + lines2, labels1 + labels2, loc = 'upper left')

plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```
